#Would you use "blue" or "circle"? (Frank & Goodman, 2012)

Lexicon, M, defining the relations between signals referents where rows represent the 4 possible signals: blue, green, circle, square and columns the 3 possible referents: blue square, blue circle, green square

Screen Shot 2022-01-04 at 2.42.19 PM.png

```python
M = [[1, 1, 0],  # blue
     [0, 0, 1],  # green
     [0, 1, 0],  # circle
     [1, 0, 1]]  # square
print("M = ", M)
```

###Literal speaker, S0

```python
import numpy as np

S0 = M / np.sum(M, axis=0);
print("S0 = ", S0)

S0_bluecircle = S0[:,1]; # preferred signal(s) for referring to the blue circle (note, 50-50)
print("S0_bluecircle = ", S0_bluecircle)
```

###Literal listener, L0

```python
L0 = M / np.sum(M,axis=1)[:,None] # probability distribution over referents given possible signals
print("L0 = ", L0)
```

###Pragmatic speaker, S1

```python
S1 = L0 / np.sum(L0, axis=0); # taking into account a literal listener
print("S1 = ", S1)

S1_bluecircle = S1[:,1]; # preferred signal(s) for referring to the blue circle
print("S1_bluecircle = ", S1_bluecircle)
```

###Confirm: Pragmatic listener, L1

```python
L1 = S1 / np.sum(S1, axis=1)[:,None] # as L0, but taking into account a pragmatic speaker
print("L1 = ", L1)

L1_circle = S1[2,:] # confirm most likely referent for circle
print("L1_circle = ", L1_circle)
```
